```python
import numpy as np
from scipy.io import loadmat
results=loadmat('yalefaces.mat')
X=results['X']
```

a.

```python
import matplotlib.pyplot as plt
```

```python
m=100
Xm=X[:, 0:m]
C = np.corrcoef(Xm.T)
print(Xm.shape)
print(C.shape)
print(C)
plt.pcolor(C)
plt.title("100 ✕ 100 Correlation Matrix")
```

b.

```python
max = [0, 1]
min = [0, 0]

for i in range(len(C)):
  for j in range (len(C)):
    if (i != j):
      if (C[i, j] > C[max[0], max[1]]):
        max = [i, j]
      if (C[i, j] < C[min[0], min[1]]):
        min = [i, j]
#max = np.unravel_index(C.argmax(), C.shape)
print(max)

#min = np.unravel_index(C.argmin(), C.shape)
print(min)
```

```python
fig = plt.figure(figsize=(10, 5))
plt.suptitle("Most Highly Correlated Faces: 6, 63")

Xp = X[:, 5].reshape((32, 32), order='F')
uplot = Xp.T
ax = fig.add_subplot(1, 2, 1)
ax.imshow(uplot)

Xp = X[:, 62].reshape((32, 32), order='F')
uplot = Xp.T
ax = fig.add_subplot(1, 2, 2)
ax.imshow(uplot)
```

```python
fig = plt.figure(figsize=(10, 5))
plt.suptitle("Least Highly Correlated Faces: 16, 82")

Xp = X[:, 15].reshape((32, 32), order='F')
uplot = Xp.T
ax = fig.add_subplot(1, 2, 1)
ax.imshow(uplot)

Xp = X[:, 81].reshape((32, 32), order='F')
uplot = Xp.T
ax = fig.add_subplot(1, 2, 2)
ax.imshow(uplot)
```

```python
m=10
indexes = np.array([1, 313, 512, 5, 2400, 113, 1024, 87, 314, 2005]) - 1
Xm=X[:, indexes]
C=np.corrcoef(Xm.T)
print(Xm.shape)
print(C.shape)
print(C)
plt.pcolor(C)
plt.title("10 ✕ 10 Correlation Matrix")
```

```python
max = [0, 1]
min = [0, 0]

for i in range(len(C)):
  for j in range (len(C)):
    if (i != j):
      if (C[i, j] > C[max[0], max[1]]):
        max = [i, j]
      if (C[i, j] < C[min[0], min[1]]):
        min = [i, j]
#max = np.unravel_index(C.argmax(), C.shape)
print(max)

#min = np.unravel_index(C.argmin(), C.shape)
print(min)
```

```python
print("Most Correlated: " + str(indexes[max[0]] + 1) + ", " + str(indexes[max[1]] + 1))
print("Least Correlated: " + str(indexes[min[0]] + 1) + ", " + str(indexes[min[1]] + 1))
```

```python
from scipy import linalg
```

```python
Y = np.matmul(X, X.T)
w, v = linalg.eig(Y)

for i in range(6):
  # Adds one so we can pretend that we are in Matlab
  print("Eigenvector " + str(i + 1) + ": " + str(v[i]))
  print("Eigenvalue " + str(i + 1) + ": " + str(w[i]))
```
